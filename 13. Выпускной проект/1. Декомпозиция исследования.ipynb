# Декомпозиция исследования

## Цель исследования

На основе истории покупок определить сегменты покупателей с целью последующего улучшения персонализированных предложений для покупателей.

## Предварительные гипотезы

Покупатели могут быть сегментированы:
- по средней стоимости корзины;
- по среднему чеку;
- по частоте оформляемых заказов или же среднему количеству оформляемых заказов за период;
- по давности оформления последнего заказа;
- по кумулятивному доходу, полученному с покупателя, за весь период с момента первой покупки;
- по категориям приобретаемых товаров;

## Основные этапы исследования

1. **Предобработка данных**
    - импорт библиотек
    - импортировать таблицы
    - вывести первые строки таблиц
    - изучить общую информацию о датасетах
    - проверить на пропуски в данных
    - проверить на наличие дублей
    - проверить корректность типов данных и преобразовать типы при необходимости
    
    
2. **Исследовательский анализ данных (EDA)**
    - изучить основные статистические показатели методом `describe()`
    - построить графики распределения
    - изучить графики "с усами"
    - сделать выводы на основании полученных данных
    
    
3. **Сегментация по средней стоимости корзины**
    - для каждого покупателя определить среднюю стоимость корзины
    - построить график распределения средней стоимости корзины по всем покупателям
    - вывести статистические показатели методом `describe()`
    - построить "ящик с усами"
    - определить границы для разделения сегментов по средней стоимости корзины
    - создать колонку с указанием сегмента для каждого покупателя
    - сделать выводы для сегментации по средней стоимости корзины
    
    
4. **Сегментация по среднему чеку**
    - для каждого покупателя определить средний чек
    - построить график распределения среднего чека по всем покупателям
    - вывести статистические показатели методом `describe()`
    - построить "ящик с усами"
    - определить границы для разделения сегментов по среднему чеку
    - создать колонку с указанием сегмента для каждого покупателя
    - сделать выводы для сегментации по среднему чеку
    
    
5. **Сегментация по частоте оформляемых заказов**
    - определить для каждого клиента среднее количество заказов за неделю, месяц и год
    - построить график распределения среднего количества заказов по всем покупателям для каждого периода
    - вывести статистические показатели методом `describe()` для каждого периода
    - построить "ящик с усами" для каждого периода
    - определить период и границы для разделения сегментов по среднему количеству заказов
    - создать колонку с указанием сегмента для каждого покупателя
    - сделать выводы для сегментации по среднему количеству заказов
    
    
6. **Сегментация по давности последнего оформленного заказа**
    - определить дату последнего оформленного заказа для каждого клиента
    - определить количество дней прошедших с момента последнего оформленного заказа
    - построить график распределения
    - вывести статистические показатели методом `describe()`
    - построить "ящик с усами" для каждого периода
    - определить границы для разделения сегментов по степени давности оформленного заказа
    - создать колонку с указанием сегмента для каждого покупателя
    - сделать выводы для сегментации по давности последнего оформленного заказа
    
    
7. **Сегментация по кумулятивному доходу, полученному с покупателя**
    - посчитать совокупный доход за весь период по каждому пользователю
    - построить график распределения
    - вывести статистические показатели методом `describe()`
    - построить "ящик с усами" для каждого периода
    - определить границы для разделения сегментов по кумулятивному доходу
    - создать колонку с указанием сегмента для каждого покупателя
    - сделать выводы для сегментации по кумулятивному доходу
    

8. **Сегментация по категориям приобретаемых товаров**
    - провести лемматизацию для наименований товаров
    - сгруппировать леммы и посчитать частоту для каждой
    - подготовить категории товаров на основе таблицы с леммами
    - определить для каждого товара категорию и записать ее в новую колонку таблицы
    - определяем самые популярные категории товаров по количеству продаж
    - определяем самые популярные категории товаров по объему продаж
    - построим диаграму рассеяния для средних значений объема продаж и количества продаж в разрезе категорий товаров
    - сделать выводы для сегментации по категориям товаров
